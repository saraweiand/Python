```
#---------- VARIÁVEIS GLOBAIS ----------
lista_funcionario = []
id_funcionario = 0

#---------- CADASTRO ----------
def cadastrar_funcionario(id):
    print('---------- Cadastramento de Funcionários ----------')
    print('--> ID do Funcionário: {}.' .format(id))
    nome = input('Nome do Funcionário: ')
    setor = input('Setor do Funcionário: ')
    salario = int(input('Salário do Funcionário: '))
    dicionario_funcionario = {'id' : id,
                              'nome' : nome,
                              'setor' : setor,
                              'salario' : salario}
    lista_funcionario.append(dicionario_funcionario.copy())

#---------- CONSULTAR ----------
def consultar_funcionario():
    print('---------- Consulta de Funcionários ----------')
    while True:
        opcao_consulta = input('********** Qual opção você deseja hoje? **********\n' +
                           '1 - Consultar Todos os Funcionários\n' +
                           '2 - Consultar Funcionário por ID\n' +
                           '3 - Consultar Funcionário(s) por Setor\n' +
                           '4 - RETORNAR\n' +
                           '>> ')
        if (opcao_consulta == '1'):
            print('Você escolheu a opção Consultar Todos os Funcionários')
            for funcionario in lista_funcionario:
                print('-------------------------')
                for key, value in funcionario.items():
                    print('{}: {}'.format(key, value))
                print('-------------------------')
        elif (opcao_consulta == '2'):
            print('Você escolheu a opção Consultar Funcionários por ID')
            id_desejado = int(input('Qual o ID que será pesquisado? '))
            for funcionario in lista_funcionario:
                if (funcionario['id'] == id_desejado):
                    print('-------------------------')
                    for key, value in funcionario.items():
                        print('{}: {}'.format(key, value))
                    print('-------------------------')
        elif (opcao_consulta == '3'):
            print('Você escolheu a opção Consultar Funcionário(s) por Setor')
            setor_desejado = input('Qual setor será pesquisado? ')
            for funcionario in lista_funcionario:
                if (funcionario['setor'] == setor_desejado):
                    print('-------------------------')
                    for key, value in funcionario.items():
                        print('{}: {}'.format(key, value))
                    print('-------------------------')
        elif (opcao_consulta == '4'):
            return
        else:
            print('!! Você digitou uma opção inválida. Tente novamente. !!')
            continue

#---------- REMOVER ----------
def remover_funcionario():
    print('---------- Remoção de Funcionários ----------')
    remocao_desejada = int(input('Qual o ID do Funcionário que você deseja remover? '))
    for funcionario in lista_funcionario:
        if (funcionario['id'] == remocao_desejada):
            lista_funcionario.remove(funcionario)
            print('Funcionário removido!')

#---------- MAIN ----------
print('---------- Bem-vindo(s) ao Controle de Funcionários da Sara Weiand Wüst ----------')
while True:
    opcao_menu = input('********** Qual opção você deseja hoje? **********\n' +
                       '1 - Cadastrar Funcionário\n' +
                       '2 - Consultar Funcionários\n' +
                       '3 - Remover Funcionário\n' +
                       '4 - SAIR\n' +
                       '>> ')
    if (opcao_menu == '1'):
        id_funcionario += 1
        cadastrar_funcionario(id_funcionario)
    elif (opcao_menu == '2'):
        consultar_funcionario()
    elif (opcao_menu == '3'):
        remover_funcionario()
    elif (opcao_menu == '4'):
        break
    else:
        print('!! Você digitou uma opção inválida. Tente novamente. !!')
        continue
```
