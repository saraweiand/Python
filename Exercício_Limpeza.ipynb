```
print('Bem-vindo ao Programa de Serviços de Limpeza de Sara Weiand Wüst')
print('****************************************************************')

def metragem_limpeza():
    print('---------Menu 1 de 3 - Metragem Limpeza----------\n')
    while True:
        while True:
                try:
                    metragem = float(input('>> Entre com a metragem em m² necessária: '))
                except ValueError:
                    print('Valor não numérico!')
                    continue
                break
        if 30 <= metragem < 300:
            valor_metragem = 60 + 0.3 * metragem
            print('Serão necessários 5 funconários para a limpeza!\n')
        elif 300 <= metragem < 700:
            valor_metragem = 120 + 0.5 * metragem
            print('Serão necessários 10 funcionários para a limpeza!\n')
        else:
            print('A empresa não aceita metragens menores que 30m² ou maiores que 700m².')
            continue
        break
    return metragem


def tipo_limpeza():
    print('---------Menu 2 de 3 - Tipo de Limpeza----------\n')
    while True:
        print('** TIPOS DE LIMPEZA: **')
        print('B - Básica. Indicada para sujeiras semananis ou quinzenais.')
        print('C - Completa. Indicada para sujeiras antigas e/ou não rotineiras.')
        tipo = input('>> Qual o tipo escolhido(B/C)? ')
        if (tipo == 'B'):
            multiplicador = 1.00
            print('Você selecionou a limpeza BÁSICA.\n')
        elif (tipo == 'C'):
            multiplicador = 1.30
            print('Você selecionou a limpeza COMPLETA.\n')
        else:
            print('!! Opção inválida !!\n')
            continue
        break
    return multiplicador

def adiconal_limpeza():
    print('---------Menu 3 de 3 - Adiconal de Limpeza----------\n')
    adicional = 0
    while True:
        print('Você deseja mais algum adicional?\n')
        print('0 - Não desejo mais nada (encerrar)')
        print('1 - Passar 10 peças de roupas. - $10.00')
        print('2 - Limpeza de 1 Forno/Micro-ondas. - $12.00')
        print('3 - Limpeza de 1 Geladeira/Freezer - $20.00')
        opcao = int(input('>> Qual a opção escolhida? \n'))
        if (opcao == 0):
            break
        elif (opcao == 1):
            adicional += 10
        elif (opcao == 2):
            adicional += 12
        elif (opcao == 3):
            adicional += 20
        else:
            print('!! Opção inválida !!')
            continue
    return adicional

metragem = metragem_limpeza()
tipo = tipo_limpeza()
adicional = adiconal_limpeza()

total = (metragem * tipo) + adicional
print('O valor final é: {}(metragem) * {}(tipo) + {}(adicional) = ${}' .format(metragem, tipo, adicional, total))
```
